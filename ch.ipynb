Я разделил все атрибуты на 3 категории. Эту таблицу ты можешь смело вставлять в свой отчет или показывать как обоснование **«Фич-инжиниринга» (Feature Selection)**.

### Сводная таблица полезности атрибутов для графового скоринга

🟢 — **Критически важно** (Оставляем 100%, это основа графа и скоринга)
🟡 — **Опционально** (Можно оставить для контекста LLM, но для базовой ML-модели не обязательно)
🔴 — **Удаляем** (Технический шум, дубликаты или мусор, который только запутает нейросеть)

| Статус | Атрибут (Колонка) | Зачем нужно в нашей задаче (Комментарий) |
| :--- | :--- | :--- |
| 🟢 | **inn** | **Узел графа (Node ID).** Уникальный идентификатор компании. Без него графа не будет. |
| 🔴 | **ogrn** | **Удаляем (Дубликат).** ОГРН выполняет ту же функцию, что и ИНН. Два ID нам не нужны. |
| 🟢 | **region** | **Связи графа (Edges).** Англоязычное название региона. Поможет LLM генерировать связи по принципу «компании из одного региона чаще торгуют друг с другом». |
| 🔴 | **region_taxcode** | **Удаляем (Дубликат).** Числовой код региона. У нас уже есть текстовый `region`, который LLM поймет лучше. |
| 🟡 | **creation_date** | **Опционально.** Дата создания. Может быть полезна, но числовой возраст (`age`) для ML-модели гораздо удобнее. |
| 🟢 | **dissolution_date** | **Супер-важно (Таргет!).** Дата ликвидации. Если она заполнена — компания "умерла" (банкрот/закрылась). Это **идеальный маркер дефолта** для обучения нашей модели! |
| 🟢 | **age** | **Фича скоринга.** Возраст в годах. Классический фактор риска: стартапы банкротятся чаще, чем "старички". |
| 🔴 | **eligible**, **exemption_criteria** | **Удаляем.** Бюрократические отметки о том, должна ли компания сдавать отчетность. Нам это не важно. |
| 🟢 | **financial** | **Для фильтрации.** Отметка "финансовая организация" (банки, страховые). Нам нужно **удалить** такие компании из выборки, так как их скорят по другим законам. |
| 🔴 | **filed**, **imputed**, **articulated**, **totals_adjustment**, **tbc_add** | **Удаляем.** Это технические флаги исследователей из ИПП (как они чистили данные). Для кредитного скоринга это информационный шум. |
| 🟡 | **simplified** | **Контекст для LLM.** Флаг "упрощенка". Объясняет, почему у компании половина финансовых строк пустые. Можно передавать в промпт. |
| 🟢 | **outlier** | **Для фильтрации.** Отметка о неправдоподобности выручки. **Удаляем** такие компании на старте, чтобы не обучать модель на сломанных данных. |
| 🟢 | **okved**, **okved_section** | **Связи графа (Edges).** Отрасль компании. Это **ключ к генерации транзакций**. LLM поймет, что строители (`okved_section = F`) покупают цемент у торговцев, а не у салонов красоты. |
| 🔴 | **okpo, okopf, okogu, okfc, oktmo** | **Удаляем.** Формы собственности и муниципальные коды. Бюрократия, не влияющая на вероятность дефолта. |
| 🟢 | **lon, lat** | **Связи графа (Edges).** Координаты! Это просто бомба для графа. Можно вычислять физическое расстояние между узлами и строить "пространственные" ребра (spatial edges). |
| 🔴 | **geocoding_quality** | **Удаляем.** Техническая отметка о точности координат. |
| 🟢 | **line_XXXX** *(очищенные)* | **Фичи скоринга.** Те самые ~30 строк баланса, которые мы отфильтровали на прошлом шаге (выручка `2110`, прибыль `2400`, активы `1600` и т.д.). Это финансовое здоровье компании. |


### Главный вывод из этой таблицы:

Смотри, как красиво теперь выстраивается архитектура твоей курсовой работы. У тебя данные четко делятся на три логических блока:

1.  **Целевая переменная (Target / Label):**
    Мы можем вычислить дефолт! Если `dissolution_date` не пустая ИЛИ `line_2400` (чистая прибыль) отрицательная несколько лет подряд — это проблемный заемщик (Дефолт = 1).
2.  **Топология графа (Откуда берутся связи):**
    Мы дадим LLM `okved` (отрасль) и географию (`region`, `lat`, `lon`). Промпт будет звучать так: *"Найди в этом радиусе логичных контрагентов для строительной фирмы и сгенерируй между ними денежные переводы"*.
3.  **Признаки узлов (Node Features):**
    Возраст (`age`) и очищенные финансовые показатели (`line_XXXX`). Это то, на чем будет обучаться нейросеть (GNN или CatBoost).

````python
import pandas as pd

data = {
    'inn':['111', '222', '333'],
    'region': ['Москва', 'Москва', 'Татарстан'],
    'okved_section': ['Строительство', 'Строительство', 'Торговля розничная'],
    'age': [5, 6, 2]
}
df = pd.DataFrame(data)

def create_text_profile(row):
    return f"Компания находится в регионе {row['region']}. Основная отрасль: {row['okved_section']}. Возраст компании: {row['age']} лет."

df['text_profile'] = df.apply(create_text_profile, axis=1)

for text in df['text_profile']:
    print(text)
````

````python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('Qwen/Qwen3-Embedding-4B')

embeddings = model.encode(df['text_profile'].tolist())

print(f"Размерность вектора: {embeddings.shape}") 
````

````python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

similarity_matrix = cosine_similarity(embeddings)

print("Похожесть компании 1 (Строители Москва) и 2 (Строители Москва):", similarity_matrix[0][1])
print("Похожесть компании 1 (Строители Москва) и 3 (Торгаши Татарстан):", similarity_matrix[0][2])
````

````python
OKVED_MAPPING = {
    "A": "Сельское хозяйство и охота",
    "B": "Добыча полезных ископаемых",
    "C": "Обрабатывающие производства",
    "D": "Энергетика и газоснабжение",
    "E": "Водоснабжение и утилизация отходов",
    "F": "Строительство",
    "G": "Торговля и ремонт автотранспорта",
    "H": "Транспортировка и хранение",
    "I": "Гостиницы и общепит",
    "J": "IT, информация и связь",
    "K": "Финансы и страхование", 
    "L": "Операции с недвижимостью",
    "M": "Профессиональная и научная деятельность",
    "N": "Административная деятельность",
    "O": "Государственное управление",
    "P": "Образование",
    "Q": "Здравоохранение и социальные услуги",
    "R": "Культура, спорт и развлечения",
    "S": "Предоставление прочих услуг",
    "T": "Деятельность домашних хозяйств",
    "U": "Экстерриториальные организации"
}
````

````python
def prepare_json(raw_json_data):
    """
    Принимает "сырой" JSON (список словарей) с полными данными компаний.
    Возвращает очищенный список словарей с INN и строкой для эмбеддера.
    """
    # Если на вход пришла строка, парсим ее в объекты Python
    if isinstance(raw_json_data, str):
        data = json.loads(raw_json_data)
    else:
        data = raw_json_data
        
    allowed_keys = {'region', 'okved_section', 'age'}
    processed_data =[]

    for company in data:
        # 1. ФИЛЬТРАЦИЯ (Удаляем мусорные компании целиком)
        # Если компания финансовая или имеет аномалии - пропускаем её
        if company.get('financial') in [True, 'True', 1, '1']:
            continue
        if company.get('outlier') in [True, 'True', 1, '1']:
            continue
            
        # 2. СБОРКА ЧИСТОГО ПРОФИЛЯ
        clean_profile = {}
        for key, value in company.items():
            if pd.isna(value) or value == "": # pd.isna лучше обрабатывает пустые значения из pandas
                continue
                
            if key in allowed_keys: # or str(key).startswith('line_'):
                # МЕНЯЕМ БУКВУ НА ТЕКСТ ДЛЯ ОКВЭД
                if key == 'okved_section':
                    # Если буква есть в словаре - берем текст, если нет - оставляем как было
                    clean_profile[key] = OKVED_MAPPING.get(value, value)
                else:
                    clean_profile[key] = value
                
        # 3. УПАКОВКА В JSON-СТРОКУ
        # Упаковываем только если профиль не оказался пустым
        if clean_profile:
            # separators=(',', ':') сжимает строку, убирая пробелы
            dense_json_string = json.dumps(clean_profile, ensure_ascii=False, separators=(',', ':'))
            
            # Сохраняем ИНН для связи с графом и готовый текст
            processed_data.append({
                "inn": company.get("inn", "unknown"),
                "text_for_embedding": dense_json_string
            })
            
    return processed_data
````

### новый заход - больше лет

Это твой «джентльменский набор» для любого кредитного аналитика. Эти пять показателей позволяют рассчитать почти все основные коэффициенты (ликвидности, рентабельности, финансовой устойчивости).

Вот их расшифровка простым языком, чтобы ты могла объяснить это любому:

### 1. `line_2110`: Выручка (Revenue)
*   **Что это:** Общий объем денег, который компания получила от продажи своих товаров или услуг за год.
*   **Зачем в скоринге:** Главный индикатор «масштаба» бизнеса. Если выручка падает из года в год — это первый сигнал, что компания теряет рынок и скоро не сможет платить по долгам.

### 2. `line_2400`: Чистая прибыль (Net Profit)
*   **Что это:** Выручка за вычетом всех расходов (зарплаты, аренда, налоги, закупки). То, что осталось в кармане у владельца.
*   **Зачем в скоринге:** Самый очевидный маркер. Прибыль — это деньги, из которых компания будет возвращать кредит. Убыток (`line_2400 < 0`) — это уже «красная зона» риска.

### 3. `line_1300`: Итого Капитал (Total Equity)
*   **Что это:** Собственные средства компании (то, что останется, если продать всё имущество и раздать все долги).
*   **Зачем в скоринге:** Твоя «подушка безопасности». Если у компании огромные долги, но огромный собственный капитал — она устойчива. Если капитал близок к нулю или отрицательный — компания держится только на заемных деньгах, она как карточный домик.

### 4. `line_1500`: Краткосрочные обязательства (Short-term Liabilities)
*   **Что это:** Все долги компании, которые нужно погасить **в течение ближайших 12 месяцев**.
*   **Зачем в скоринге:** Критически важная цифра. Если `line_1500` больше, чем все активы компании, то компания технически банкрот прямо сейчас — она не может расплатиться по долгам в срок.

### 5. `line_1600`: Итого Активы (Total Assets)
*   **Что это:** Все, чем владеет компания (деньги на счетах, станки, товары на складе, офисы).
*   **Зачем в скоринге:** Показывает размер компании. Активы должны быть больше, чем долги (`line_1500`).

---

### Как эти 5 строк работают вместе (для LLM):

Когда ты отдашь эти 5 цифр в промпт, LLM (Qwen3-8B) внутри своей "головы" моментально сделает следующие вычисления:

1.  **Ликвидность:** `line_1600 / line_1500`. Если меньше 1 — риск дефолта 100%.
2.  **Рентабельность:** `line_2400 / line_2110`. Если отрицательная — бизнес работает впустую.
3.  **Финансовая независимость:** `line_1300 / line_1600`. Если доля собственного капитала (капитала) меньше 20% — бизнес зависит от кредиторов (то есть от банка).

**Теперь ты понимаешь, почему я выбрал именно эти пять!** Они закрывают все главные риски:
*   Масштаб (`2110`)
*   Эффективность (`2400`)
*   Устойчивость (`1300`)
*   Платежеспособность (`1500`)
*   Ресурсная база (`1600`)

````python
import pandas as pd
import numpy as np
from tqdm import tqdm

usecols =['inn', 'region', 'okved_section', 'age', 'dissolution_date', 
           'line_2110', 'line_2400', 'line_1300', 'line_1500', 'line_1600', 
           'financial', 'outlier']

TARGET_REGIONS =['moscow city', 'st. petersburg', 'tatarstan', 'sverdlovsk', 'krasnodar']
# TARGET_OKVEDS =['C', 'F', 'G', 'J', 'M']

# --- ЭТАП 1: СОБИРАЕМ 10 000 ИНН ИЗ 2020 ГОДА (чтобы поймать и живых, и мертвых) ---
print("1. Формируем ядро из 10 000 ИНН (Базовый год: 2020)...")
core_inns = set()
chunks_2020 =[]

for chunk in pd.read_csv('./data_rfsd/csv/data_150_rfsd_2020_v20251010.csv', sep=';', usecols=usecols, chunksize=500000):
    c = chunk[(chunk['financial'] != 1) & (chunk['outlier'] != 1) & (chunk['line_2110'] > 0)].copy()
    c = c[(c['region'].isin(TARGET_REGIONS))] # & (c['okved_section'].isin(TARGET_OKVEDS))]
    chunks_2020.append(c)

df_2020_base = pd.concat(chunks_2020, ignore_index=True)
df_2020_sample = df_2020_base.sample(n=min(10000, len(df_2020_base)), random_state=42)
target_inns = set(df_2020_sample['inn'])
del df_2020_base, chunks_2020 # Чистим память
print(f"✅ Ядро собрано. Уникальных ИНН: {len(target_inns)}")


# --- ЭТАП 2: СОБИРАЕМ ИСТОРИЮ ЗА ВСЕ 8 ЛЕТ ДЛЯ ЭТИХ ИНН ---
print("\n2. Собираем историю с 2017 по 2024 год...")
years =[2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
all_history =[]

for year in tqdm(years, desc="Обработка годов"):
    filepath = f'./data_rfsd/csv/data_150_rfsd_{year}_v20251010.csv'
    try:
        for chunk in pd.read_csv(filepath, sep=';', usecols=usecols, chunksize=500000):
            # ВЫТАСКИВАЕМ ТОЛЬКО НАШИ 10 000 КОМПАНИЙ! (Это жрет 0 памяти)
            c = chunk[chunk['inn'].isin(target_inns)].copy()
            c['year'] = year
            all_history.append(c)
    except FileNotFoundError:
        print(f"Файл за {year} не найден, пропускаем...")

df_full_history = pd.concat(all_history, ignore_index=True)


# --- ЭТАП 3: АГРЕГАЦИЯ 8-ЛЕТНЕЙ ИСТОРИИ В ОДНУ СТРОКУ ---
print("\n3. Считаем многолетние тренды (2017-2024)...")
df_full_history = df_full_history.sort_values(by=['inn', 'year'])

def aggregate_8_years(group):
    first_record = group.iloc[0]
    last_record = group.iloc[-1]
    
    # БЕЗОПАСНОЕ ИЗВЛЕЧЕНИЕ ДЕНЕГ (Заменяем NaN на 0)
    rev_first = first_record['line_2110'] if pd.notna(first_record['line_2110']) else 0
    rev_last = last_record['line_2110'] if pd.notna(last_record['line_2110']) else 0
    
    prof_first = first_record['line_2400'] if pd.notna(first_record['line_2400']) else 0
    prof_last = last_record['line_2400'] if pd.notna(last_record['line_2400']) else 0
    
    years_active = len(group)
    
    # СЧИТАЕМ ОТНОСИТЕЛЬНЫЙ ТРЕНД (%)
    # Если начальная выручка была 0, а стала 100, считаем это ростом на 100% (условно 1.0)
    if rev_first > 0:
        rev_trend_pct = (rev_last - rev_first) / rev_first
    else:
        rev_trend_pct = 1.0 if rev_last > 0 else 0.0
        
    # Абсолютный среднегодовой тренд прибыли (т.к. проценты с убытками считать математически некорректно)
    prof_trend_per_year = (prof_last - prof_first) / years_active

    return pd.Series({
        'region': last_record['region'],
        'okved_section': last_record['okved_section'],
        'max_age': last_record['age'],
        'years_in_dataset': years_active, 
        
        'real_default': 1 if pd.notna(last_record['dissolution_date']) else 0,
        
        'last_rev': rev_last,
        'last_prof': prof_last,
        'last_cap': last_record['line_1300'] if pd.notna(last_record['line_1300']) else 0,
        'last_debt': last_record['line_1500'] if pd.notna(last_record['line_1500']) else 0,
        'last_assets': last_record['line_1600'] if pd.notna(last_record['line_1600']) else 0,
        
        # НОВЫЕ ПРАВИЛЬНЫЕ ТРЕНДЫ (Без NaN)
        'revenue_trend_pct': round(rev_trend_pct, 4), # Например, 0.15 означает рост на 15%
        'profit_trend_per_year': round(prof_trend_per_year, 2)
    })

final_df = df_full_history.groupby('inn').apply(aggregate_8_years).reset_index()

print(f"\n✅ ГОТОВО! Итоговый датасет: {len(final_df)} компаний.")
print(f"Банкротов в выборке: {final_df['real_default'].sum()}")
display(final_df.head(3))
````

````python
OKVED_MAPPING = {
    "A": "Сельское хозяйство и охота",
    "B": "Добыча полезных ископаемых",
    "C": "Обрабатывающие производства",
    "D": "Энергетика и газоснабжение",
    "E": "Водоснабжение и утилизация отходов",
    "F": "Строительство",
    "G": "Торговля и ремонт автотранспорта",
    "H": "Транспортировка и хранение",
    "I": "Гостиницы и общепит",
    "J": "IT, информация и связь",
    "K": "Финансы и страхование", 
    "L": "Операции с недвижимостью",
    "M": "Профессиональная и научная деятельность",
    "N": "Административная деятельность",
    "O": "Государственное управление",
    "P": "Образование",
    "Q": "Здравоохранение и социальные услуги",
    "R": "Культура, спорт и развлечения",
    "S": "Предоставление прочих услуг",
    "T": "Деятельность домашних хозяйств",
    "U": "Экстерриториальные организации"
}
````

````python
def prepare_json(raw_json_data):
    if isinstance(raw_json_data, str):
        data = json.loads(raw_json_data)
    else:
        data = raw_json_data
        
    # ВНИМАНИЕ: Здесь ТОЛЬКО бизнес-фичи! Никаких дат ликвидаций!
    allowed_keys = {
        'region', 'okved_section', 'max_age', 
        'last_rev', 'last_prof', 'last_cap', 'last_debt', 'last_assets',
        'revenue_trend_pct', 'profit_trend_per_year'
    }
    
    processed_data =[]

    for company in data:
        # СБОРКА ЧИСТОГО ПРОФИЛЯ ДЛЯ ЭМБЕДДЕРА
        clean_profile = {}
        for key, value in company.items():
            if pd.isna(value) or value == "": 
                continue
                
            if key in allowed_keys:
                if key == 'okved_section':
                    clean_profile[key] = OKVED_MAPPING.get(value, value)
                else:
                    if isinstance(value, float):
                        clean_profile[key] = round(value, 4)
                    else:
                        clean_profile[key] = value
                
        # УПАКОВКА
        if clean_profile:
            dense_json_string = json.dumps(clean_profile, ensure_ascii=False, separators=(',', ':'))
            
            # Сохраняем ИНН, прокси-таргет (для будущей сверки) и чистый текст (для NLP)
            processed_data.append({
                "inn": company.get("inn"),
                "proxy_target": company.get("real_default", 0), # Наш ориентировочный таргет
                "text_for_embedding": dense_json_string
            })
            
    return processed_data
````

````python
import numpy as np
import networkx as nx
from sklearn.neighbors import NearestNeighbors

embeddings = np.load('embeddings_10000.npy')
print(f"Эмбеддинги загружены. Размер: {embeddings.shape}")

# 1. СТРОИМ ЧИСТЫЙ k-NN ГРАФ
# Ищем 6 ближайших точек (1-я это сама компания, её отбросим, останется 5 реальных соседей)
print("Ищем 5 ближайших соседей для каждой компании...")
knn = NearestNeighbors(n_neighbors=6, metric='cosine')
knn.fit(embeddings)
distances, indices = knn.kneighbors(embeddings)

G = nx.Graph()
G.add_nodes_from(range(len(embeddings)))

print("Сшиваем граф...")
for i in range(len(indices)):
    for j in indices[i][1:]: # Пропускаем индекс 0 (самого себя)
        G.add_edge(i, j)

# --- ПРОВЕРКА ИДЕАЛА ---
degrees = [degree for node, degree in G.degree()]
isolates = list(nx.isolates(G))

print(f"\n✅ ГОТОВО!")
print(f"Узлов: {G.number_of_nodes()}")
print(f"Связей: {G.number_of_edges()}")
print(f"Изолированных вершин: {len(isolates)}")
print(f"Среднее число связей: {np.mean(degrees):.1f}")
print(f"Максимальное число связей (ХАБ): {max(degrees)}")
````
